```python
# Create index of 100 students and save it in a list 
index_student = [idx+1 for idx in range(len(gender))]
#print(index_student) #นักเรียน 100 คน

#ดึงค่าคะแนนของนักเรียนทั้ง 3 วิชามาจากDataFrameของPandas
math_score = csv_data['math score']
reading_score = csv_data['reading score']
writing_score = csv_data['writing score']
```

```python
#เรียงค่าคะแนนวิชาคณิตศาตร์ และวิชาการอ่านจากน้อยไปหามาก
#วิชาคณิตศาสตร์ 
#Write code below please.
sort_math_score=list(math_score.sort_values())

#วิชาการอ่าน
#Write code below please.
sort_reading_score=list(reading_score.sort_values())
```

```python
plt.figure(figsize=(10,6))
# Write code below please. #
studentCount=range(10)
plt.plot(studentCount,sort_math_score[:10],color="red",label="Math")
plt.plot(studentCount,sort_reading_score[:10],color="blue",label="Reading")
plt.xticks(studentCount,studentCount)
plt.legend(loc="best")
############################
plt.title('Line chart comparison on Math and Reading score of 10 students')
plt.ylabel('Score')
plt.xlabel('Individual student')
plt.show()
```

```python
#Write code below please.
# mean_math = math_score.sum()/len(math_score)
# mean_writing = writing_score.sum()/len(writing_score)
# mean_reading = reading_score.sum()/len(reading_score)

# sort_math_score=list(math_score.sort_values())
# sort_writing_score=list(writing_score.sort_values())
# sort_reading_score=list(reading_score.sort_values())

# median_math = (sort_math_score[(len(sort_math_score)+1)//2]+sort_math_score[(len(sort_math_score)+2)//2])/2
# median_reading = (sort_reading_score[(len(sort_reading_score)+1)//2]+sort_reading_score[(len(sort_reading_score)+2)//2])/2
# median_writing = (sort_writing_score[(len(sort_writing_score)+1)//2]+sort_writing_score[(len(sort_writing_score)+2)//2])/2

mean_math = math_score.mean()
mean_writing = writing_score.mean()
mean_reading = reading_score.mean()

median_math = math_score.median()
median_reading = writing_score.median()
median_writing = reading_score.median()
```

```python
#Plot multiple bar graph
#Write code below please.
plt.figure(figsize=(20,10))
width=0.4
index_number=[i+(width/2) for i in range(3)]
index_mean=[i for i in range(3)]
index_median=[i+width for i in range(3)]

plt.bar(index_mean,[mean_math,mean_reading,mean_writing],width,label="Mean score",color="violet")
plt.bar(index_median,[median_math,median_reading,median_writing],width,label="Median score",color="lightblue")
plt.ylabel("Scores")
plt.title("Scores by group and gender")
plt.xticks(index_number,("Math","Reading","Writing"))
plt.legend(loc="best")
plt.show()
```
